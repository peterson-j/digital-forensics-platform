```python
# Cell 1: Install required packages and define a basic IngestionEngine for file type detection

!pip install python-magic python-magic-bin pandas

import os
import magic
import pandas as pd

class IngestionEngine:
    """
    Minimal version: detects file type only.
    """
    def __init__(self):
        print("IngestionEngine initialized (file detection only).")

    def detect_file_type(self, file_path):
        """
        Detect MIME type using python-magic.
        Returns a string: 'image', 'video', 'audio', 'text', 'tabular', or 'unknown'.
        """
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"File not found: {file_path}")

        mime = magic.from_file(file_path, mime=True)
        if mime.startswith('image'):
            return 'image'
        elif mime.startswith('video'):
            return 'video'
        elif mime.startswith('audio'):
            return 'audio'
        elif mime.startswith('text') or mime in ['application/json', 'application/pdf']:
            return 'text'
        elif mime in ['text/csv', 'application/vnd.ms-excel', 'application/vnd.openxmlformats-officedocument.spreadsheetml.sheet']:
            return 'tabular'
        else:
            return 'unknown'

# Test the engine with a sample file (you'll need to upload a file first)
# For now, we'll just create the engine and you can test later.
engine = IngestionEngine()
print("Ready to detect file types. Use engine.detect_file_type('your_file_path')")
```

```python
# ============================================
# DATA INGESTION LAYER: File Type Detection
# ============================================

# 1. Install required system libraries and Python packages
!apt-get update -qq && apt-get install -y libmagic1
!pip install python-magic pandas

# 2. Imports
import os
import magic
import pandas as pd
from google.colab import files

# 3. IngestionEngine class with broad category detection
class IngestionEngine:
    """
    Minimal version: detects file type and categorizes into broad groups.
    """
    def __init__(self):
        print("IngestionEngine initialized (file detection only).")

    def detect_file_type(self, file_path):
        """
        Detect MIME type using python-magic.
        Returns a broad category: 'image', 'video', 'audio', 'text', 'tabular', or 'unknown'.
        """
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"File not found: {file_path}")

        mime = magic.from_file(file_path, mime=True)

        # Image types
        if mime.startswith('image'):
            return 'image'
        # Video types
        elif mime.startswith('video'):
            return 'video'
        # Audio types
        elif mime.startswith('audio'):
            return 'audio'
        # Plain text and common document formats (treated as text)
        elif (mime.startswith('text') or
              mime in ['application/pdf',
                       'application/msword',
                       'application/vnd.openxmlformats-officedocument.wordprocessingml.document',
                       'application/vnd.oasis.opendocument.text',
                       'application/rtf',
                       'application/json']):
            return 'text'
        # Tabular (spreadsheets)
        elif mime in ['text/csv',
                      'application/vnd.ms-excel',
                      'application/vnd.openxmlformats-officedocument.spreadsheetml.sheet']:
            return 'tabular'
        else:
            return 'unknown'

# 4. Create an instance of the engine
engine = IngestionEngine()

# 5. Upload a file and test detection
print("\nPlease upload a file to test file type detection.")
uploaded = files.upload()

for filename, content in uploaded.items():
    # Save the uploaded file to the current directory
    with open(filename, 'wb') as f:
        f.write(content)
    print(f"\nUploaded: {filename}")

    # Detect file type
    file_type = engine.detect_file_type(filename)
    print(f"Detected category: {file_type}")

print("\nDone. You can upload more files by running this cell again.")
```
